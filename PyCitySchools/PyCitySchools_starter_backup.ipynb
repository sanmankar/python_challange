## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Create the data frame to hold the district summary results
cols = ['Tot_Num_Schools','Tot_Num_Students','Tot_Budget','Avg_Math_Score','Avg_Reading_Score','Overall_Avg_Score',\
     'Overall_Math_Pass_Pct','Overall_Reading_Pass_Pct']
v_index = {0}

district_summary = pd.DataFrame(index=v_index, columns=cols)
district_summary.columns

# Assignig value from a data frame to a variable. This is just to look for an alternative way.
x = len(school_data_complete['School_ID'].unique())

# Calculating and storing total number of schools
district_summary['Tot_Num_Schools'] = len(school_data_complete['School_ID'].unique())
# Calculating ans storing total number of students
district_summary['Tot_Num_Students'] = school_data_complete['student_name'].count()
# Calculating the total budget
district_summary['Tot_Budget'] = school_data['budget'].sum()
district_summary['Tot_Budget'] = district_summary['Tot_Budget'].astype(float).map("${:,.2f}".format)
# Calculating Average Math Score
district_summary['Avg_Math_Score'] = school_data_complete['math_score'].mean()
district_summary['Avg_Math_Score'] = district_summary['Avg_Math_Score'].round()
# Calculating Average Reading Score
district_summary['Avg_Reading_Score'] = school_data_complete['reading_score'].mean()
district_summary['Avg_Reading_Score'] = district_summary['Avg_Reading_Score'].round()
# Calculating Average Reading Score
district_summary['Overall_Avg_Score'] = (district_summary['Avg_Math_Score'] + district_summary['Avg_Reading_Score'])/2
district_summary['Overall_Avg_Score'] = district_summary['Overall_Avg_Score'].round()
# Calculating Overall Math Pass Percentage ( pass score > 70 )
district_summary['Overall_Math_Pass_Pct'] = (school_data_complete.loc[school_data_complete['math_score'].astype(float) >= 70]['Student_ID'].count()/district_summary['Tot_Num_Students']) *100
district_summary['Overall_Math_Pass_Pct'] = district_summary['Overall_Math_Pass_Pct'].astype(float).map("%{:,.2f}".format)
# Calculating Overall Reading Pass Percentage ( pass score > 70 )
district_summary['Overall_Reading_Pass_Pct'] = (school_data_complete.loc[school_data_complete['reading_score'].astype(float) >= 70]['Student_ID'].count()/district_summary['Tot_Num_Students']) *100
district_summary['Overall_Reading_Pass_Pct'] = district_summary['Overall_Reading_Pass_Pct'].astype(float).map("%{:,.2f}".format)

district_summary
```

```python
v_school_summary = {}
v_school_summary = school_data_complete.groupby(['school_name','type'])['student_name'].count()

# Create a school_summary data fraome from the three columns listed in the summary and reanme
# the student_name column to Tot_Students
school_summary = pd.DataFrame.from_dict(v_school_summary)
school_summary = school_summary.rename(columns={'student_name':'Tot_Students'})
#school_summary = school_summary['Tot_Students']


# Calculating the total budget for each school. Remember to take the school data else the buget values will be inflated
school_summary['Tot_Budget'] = school_data.groupby(['school_name','type'])['budget'].sum()
# Calculate per student budget
school_summary['Per_Stud_Budget'] = school_summary['Tot_Budget'].astype(float)/school_summary['Tot_Students']
# Calcualte average match score
school_summary['Avg_Math_Score'] = school_data_complete.groupby(['school_name','type'])['math_score'].mean()
school_summary['Avg_Math_Score'] = school_summary['Avg_Math_Score'].round()
# Calcualte average reading score
school_summary['Avg_Reading_Score'] = school_data_complete.groupby(['school_name','type'])['reading_score'].mean()
school_summary['Avg_Reading_Score'] = school_summary['Avg_Reading_Score'].round()


# Format values in dataframe
school_summary['Tot_Budget'] = school_summary['Tot_Budget'].astype(float).map("${:,.2f}".format)
school_summary['Per_Stud_Budget'] = school_summary['Per_Stud_Budget'].astype(float).map("${:,.2f}".format)


# Resetting the index to a series from 0 to len(dataframe). 
# This is needed because the prior steps create the school_name and type as index.
# You could use the command df.set_index("column_name") to reset the index to a particular column
# school_summary.reset_index(inplace=True)



school_summary
```
